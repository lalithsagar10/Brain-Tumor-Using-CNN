```python
def names(number):
    if number==0:
        return 'Its a Tumor'
    else:
        return 'No, Its not a tumor'
```

```python

model = tf.keras.models.load_model('/content/drive/MyDrive/brain tumour/brain.h5')
def classify_image(image):
  image = np.array(image)
  #x = image
  l = np.count_nonzero(image == 1)
  #image = np.array(image.resize((128,128)))
  image = image.reshape(1,128,128,3)
  res = model.predict(image)
  classification = np.where(res == np.amax(res))[1][0]
  s = area = None
  if classification == 0:
    s = np.std(res)
    area = (0.264)* math.sqrt(l)
  return names(classification), area, s,classification


image = gr.inputs.Image(shape=(128,128))

output = gr.Textbox(label="Prediction")
output_area = gr.Number(label="Area:")
output_intensity = gr.Number(label="Intesity: ")






  
gr.Interface(fn=classify_image, inputs=image, 
             outputs=[output, output_area, output_intensity ] ,
             capture_session=True).launch(debug='True')


```
